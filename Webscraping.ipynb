## USE BEAUTIFUL SOUP TO PARSE AND EXTRACT INFORMATION

```python
!pip install beautifulsoup4 --upgrade --quiet
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page_contents,'html.parser')
```

```python
job_titles = []
for h3_tag in soup.find_all('h3', class_='heading_4_5 profile'):
    job_titles.append(h3_tag.a.text)

# Print the list of job titles
print(len(job_titles))
print(job_titles)
```

```python
company_name = []
for div_tag in soup.find_all('div', class_='company_and_premium'):
    company_name.append(div_tag.a.text.strip())

# Print the list of company names
print(len(company_name))
print(company_name)
```

```python
location = []
for a_tag in soup.find_all('a', class_='location_link view_detail_button'):
    location.append(a_tag.text.strip())

# Print the list of location names
print(len(location))
print(location)
```

```python
stipend = []
for span_tag in soup.find_all('span', class_='stipend'):
    # Extract the full text from the span tag
    full_text = span_tag.text.strip()
    
    # Split the text by space
    parts = full_text.split(' ')
    
    # Check if there are at least two parts after splitting
    if len(parts) > 1:
        # Extract the amount in rupees
        amount = parts[1]  # Assumes the amount is the second part after splitting by space
        stipend.append(amount)
    else:
        stipend.append('Amount not specified')  # Or handle this case as per your requirement

# Print the list of stipend amounts
print(len(stipend))
print(stipend)
```

```python
# Find all 'a' tags with the specified class and get the value of the "href" attribute for each
all_links = []
base_url= 'https://internshala.com/'
for a_tag in soup.find_all('a', class_='btn btn-secondary view_detail_button_outline'):
    href_link = a_tag.get('href')
    all_links.append(base_url + href_link)

# Print the list of links
print(all_links)
```

```python
import pandas as pd
```

```python
jobs_dict = {
    'job_titles': job_titles,
    'company_name': company_name,
    'location': location,
    'stipend': stipend,
    'all_links': all_links
}
```

```python
jobs_df = pd.DataFrame(jobs_dict)
```
